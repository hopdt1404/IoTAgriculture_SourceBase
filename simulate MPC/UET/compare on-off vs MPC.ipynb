```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

soilMoistureFolowTimeByONOFF=pd.read_csv("soil_moisture_on_off.csv")
soilMoistureFolowTimeByMPC = pd.read_csv("soil_moisture_MPC.csv")

```

```python
sumTimeOnOff=[]
sumTimeOnOff.append(soilMoistureFolowTimeByONOFF['amount of watering'][0])
for i in range(1,len(soilMoistureFolowTimeByONOFF['amount of watering'])):
    sumTimeOnOff.append(sumTimeOnOff[i-1]+soilMoistureFolowTimeByONOFF['amount of watering'][i])
```

```python
sumTimeMPC=[]
sumTimeMPC.append(soilMoistureFolowTimeByMPC['amount of watering'][0])
for i in range(1,len(soilMoistureFolowTimeByMPC['amount of watering'])):
    sumTimeMPC.append(sumTimeMPC[i-1]+soilMoistureFolowTimeByMPC['amount of watering'][i])
```

```python
pd.to_datetime(soilMoistureFolowTimeByONOFF['datetime'])[0].hour
```

```python
figure(num=None, figsize=(8, 4), dpi=1000, facecolor='w', edgecolor='k')
plt.plot(pd.to_datetime(soilMoistureFolowTimeByONOFF['datetime']),sumTimeOnOff)
plt.plot(pd.to_datetime(soilMoistureFolowTimeByMPC['datetime']),sumTimeMPC)
plt.title('Tổng thời lượng tưới nước giữa 2 phương pháp tưới On/Off và MPC')
plt.legend(['On/Off','MPC'])
plt.xlabel('thời gian (tháng-ngày giờ)')
plt.ylabel('tổng thời lượng tưới nước (giây)')
plt.savefig('compare_sum.svg')
```

```python
figure(num=None, figsize=(8, 4), dpi=1000, facecolor='w', edgecolor='k')
plt.plot(pd.to_datetime(soilMoistureFolowTimeByONOFF['datetime']),soilMoistureFolowTimeByONOFF['soil moisture'])
plt.plot(pd.to_datetime(soilMoistureFolowTimeByMPC['datetime']),soilMoistureFolowTimeByMPC['soil moisture'])
plt.title('Sự biến thiên độ ẩm đất giữa 2 phương pháp tưới On/Off và MPC')
plt.legend(['On/Off','MPC'])
plt.xlabel('thời gian (tháng-ngày giờ)')
plt.ylabel('độ ẩm đất (%)')
plt.savefig('compare_value.svg')
```
